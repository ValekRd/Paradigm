```python
# Pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# ignore warnings
import warnings
warnings.filterwarnings('ignore')

import yapo as y
from datetime import datetime

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

#векторный формат
%config InlineBackend.figure_format = 'svg'
```

## 1. Прогружаем данные

```python
BTC = pd.read_csv('Data/BTCUSD.csv', index_col='Date', parse_dates=True)
ETH = pd.read_csv('Data/ETHUSD.csv', index_col='Date', parse_dates=True)
#удаляем более ранние данные по Bitcoin
BTC.drop(BTC.tail(len(BTC) - len(ETH)).index,inplace=True)
```

```python
BTC['Price'] = BTC['Price'].str.replace(',', '').astype(float)
ETH['Price'] = ETH['Price'].str.replace(',', '').astype(float)
```

## 2. Смотрим динамику курса

```python
figure(figsize=(11,5))
plt.plot(BTC['Price'][0::3], color='green', marker='', linestyle='-', markerfacecolor='blue', label = 'Bitcoin');
plt.grid(linestyle='--', linewidth=0.5)

#plt.legend(loc='upper center', ncol=2);
plt.legend(loc='best');        

#Подписи для осей:
#plt.xlabel('time', fontsize=10)
plt.ylabel('USD', fontsize=10);

#Диапазон осей:
#plt.xlim('2016-03-14', '2019-08-01');
#plt.ylim(0, 3000);

# Градуировка осей
#plt.yticks(range(0, 3200, 500), fontsize=12)

#Заголовок:
plt.title('Динамика курса', fontsize=16, y=1.05);
```

```python
figure(figsize=(11,5))
plt.plot(ETH['Price'][0::3], color='grey', marker='', linestyle='-', markerfacecolor='blue', label = 'Ethereum');
plt.grid(linestyle='--', linewidth=0.5)
plt.legend(loc='best');        
plt.ylabel('USD', fontsize=10);
plt.title('Динамика курса', fontsize=16, y=1.05);
```

```python
# корреляция между BTC и ETH (измеряется от -1 до 1)
BTC['Price'].corr(ETH['Price'])
```
